## Theory

A Multi-Layer Perceptron (MLP) is one of the simplest types of artificial neural networks.

It consists of:
- An input layer
- One or more hidden layers
- An output layer

Each neuron in one layer is connected to every neuron in the next layer, which is why it is also called a **Fully Connected Neural Network**.

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
# Input data
X = torch.tensor([
    [0., 0.],
    [0., 1.],
    [1., 0.],
    [1., 1.]
])

# Target labels
y = torch.tensor([
    [0.],
    [1.],
    [1.],
    [0.]
])
```

## Define the MLP Model

In PyTorch, neural networks are created by defining a class that inherits from `nn.Module`.

The model consists of:

- An input layer
- A hidden layer with ReLU activation
- An output layer

This architecture is called a **Multi-Layer Perceptron (MLP)**

```python
class SimpleMLP(nn.Module):
    def __init__(self):
        super().__init__()

        self.fc1 = nn.Linear(2, 4)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(4, 1)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
```

```python
# Create the model

model = SimpleMLP()

print(model)
```

## Define the Loss Function and Optimizer

Before training the model, we need:

- **Loss Function**: Measures how far the model's predictions are from the true values.
- **Optimizer**: Updates the model's parameters to reduce the loss.

```python
# Loss Function
criterion = nn.BCEWithLogitsLoss()

# Optimizer
optimizer = optim.SGD(model.parameters(), lr=0.1)
```

## Train the Model

The training loop performs the following steps repeatedly:

1. Make predictions (Forward Pass)
2. Compute the loss
3. Calculate gradients (Backward Pass)
4. Update model weights
5. Clear old gradients

```python
# Training Loop

epochs = 1000

for epoch in range(epochs):

    # Forward pass
    outputs = model(X)

    # Calculate loss
    loss = criterion(outputs, y)

    # Clear previous gradients
    optimizer.zero_grad()

    # Backward pass
    loss.backward()

    # Update weights
    optimizer.step()

    # Print loss every 100 epochs
    if (epoch + 1) % 100 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")
```

## Evaluate the Model

After training, we can use the trained model to make predictions on the input data.

```python
# Disable gradient calculation during evaluation
with torch.no_grad():

    outputs = model(X)

    # Convert logits to probabilities
    predictions = torch.sigmoid(outputs)

    print(predictions)
```

```python
# Convert probabilities to class labels

predicted_classes = (predictions >= 0.5).float()

print(predicted_classes)
print("Actual Labels:")
print(y)
```
