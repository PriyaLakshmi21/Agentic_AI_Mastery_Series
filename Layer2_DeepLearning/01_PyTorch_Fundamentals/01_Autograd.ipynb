# 📖 Introduction

Training a neural network requires calculating gradients for millions (or even billions) of parameters.

Computing these gradients manually is slow, error-prone, and practically impossible for large neural networks.

PyTorch solves this problem using **Autograd**, its automatic differentiation engine.

Autograd records every mathematical operation performed on tensors and builds a **computational graph**. When we call `backward()`, PyTorch applies the **chain rule** to compute gradients automatically.

These gradients are then used by optimizers such as **SGD** and **Adam** to update the model parameters and reduce the prediction error.

# ❓ Why Do We Need Autograd?

Imagine training ChatGPT.

A single modern language model contains **billions of parameters**.

If we had to manually calculate the derivative of every parameter after every prediction:

- It would take an enormous amount of time.
- The calculations would be extremely complex.
- Even a tiny mistake would produce incorrect gradients.

Autograd automates this entire process.

It computes gradients accurately and efficiently, making deep learning practical.

# 🔹 Step 1 – Create a Tensor

In PyTorch, data is stored inside **Tensors**.

To calculate gradients, we must tell PyTorch to track operations on a tensor.

We do this using:

```python
requires_grad=True
```

This tells Autograd:

> "Record every operation performed on this tensor because I will need its derivative later."

````python
# Create a tensor with gradient tracking enabled

x = torch.tensor(2.0, requires_grad=True)

print("Tensor:", x)
````

# 🔹 Step 2 – Define a Mathematical Function

Let's define the function:

\[
y = 3x^3 - 2x
\]

PyTorch will automatically remember every operation used to compute this expression.

````python
# Define the function

y = 3 * x**3 - 2 * x

print("Value of y:", y)
````

# 🔹 Step 3 – Compute the Gradient

Now we call:

```python
y.backward()
```

This tells PyTorch:

> "Compute the derivative of y with respect to x."

Autograd traverses the computational graph backwards and calculates the gradient automatically.

````python
# Compute gradients

y.backward()

print("Gradient:", x.grad)
````

# 🧮 Mathematical Verification

We defined:

\[
y = 3x^3 - 2x
\]

Differentiate with respect to x:

\[
\frac{dy}{dx} = 9x^2 - 2
\]

Substitute:

\[
x = 2
\]

\[
9(2)^2 - 2
\]

\[
9(4)-2
\]

\[
36-2=34
\]

PyTorch computed the same result.

✅ Gradient = 34

# 📌 Observations

- Autograd automatically computes derivatives.
- No manual differentiation is required.
- Gradients are stored inside the `.grad` attribute.
- The computation graph is built dynamically during execution.

# ✅ Conclusion
