# 📖 Introduction

Artificial Neural Networks are built by connecting multiple neurons together.

Each neuron performs two operations:

1. Computes a weighted sum of its inputs.
2. Passes the result through an **Activation Function**.

Without an activation function, every neuron would simply perform a linear transformation.

As a result, no matter how many layers we add, the entire neural network would behave like a single linear model.

Activation functions introduce **non-linearity**, allowing neural networks to learn complex patterns such as images, speech, and natural language.

# ❓ Why Do We Need Activation Functions?

Imagine you want to recognize handwritten digits.

The relationship between pixels and the digit is highly complex.

A straight line can never separate all possible images of digits.

If every neuron only performed linear calculations, the neural network would never learn these complex relationships.

Activation functions solve this problem by introducing **non-linearity** into the network.

This enables Deep Learning models to solve problems that traditional linear models cannot.

# 🧠 What is an Activation Function?

An Activation Function is a mathematical function applied to the output of a neuron.

It decides how much information should pass to the next layer.

Instead of allowing every value to pass unchanged, it transforms the output in a way that helps the network learn meaningful patterns.

Different activation functions behave differently, making them suitable for different types of problems.

```python
# Import required libraries

import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
# Generate input values

x = torch.linspace(-5, 5, 1000)
```

## Sigmoid Activation Function

The Sigmoid activation function maps any real-valued input to a value between **0 and 1**.

It is commonly used in the output layer of binary classification models because its output can be interpreted as a probability.

```python
# Create Sigmoid activation

sigmoid = nn.Sigmoid()

# Apply activation

y_sigmoid = sigmoid(x)

# Plot

plt.figure(figsize=(6,4))
plt.plot(x.numpy(), y_sigmoid.numpy())
plt.title("Sigmoid Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

### Observation

- Output Range: **0 to 1**
- Smooth and differentiable
- Suitable for binary classification output layers
- Suffers from the vanishing gradient problem

## Tanh Activation Function

The Tanh (Hyperbolic Tangent) activation function maps input values to the range **-1 to 1**.

Unlike Sigmoid, it is zero-centered, which often helps neural networks learn faster.

```python
# Create Tanh activation

tanh = nn.Tanh()

# Apply activation

y_tanh = tanh(x)

# Plot

plt.figure(figsize=(6,4))
plt.plot(x.numpy(), y_tanh.numpy())
plt.title("Tanh Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

### Observation

- Output Range: **-1 to 1**
- Zero-centered
- Performs better than Sigmoid in many cases
- Still suffers from the vanishing gradient problem

## ReLU Activation Function

The Rectified Linear Unit (ReLU) is the most widely used activation function in deep learning.

It returns **0** for negative inputs and returns the input itself for positive values.

ReLU is computationally efficient and helps reduce the vanishing gradient problem, making it the default choice for hidden layers in many neural networks.

```python
# Create ReLU activation

relu = nn.ReLU()

# Apply activation

y_relu = relu(x)

# Plot

plt.figure(figsize=(6,4))
plt.plot(x.numpy(), y_relu.numpy())
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

### Observation

- Output Range: **0 to ∞**
- Very fast to compute
- Does not saturate for positive values
- Most commonly used activation function for hidden layers
- Can suffer from the **Dying ReLU Problem**

## Leaky ReLU Activation Function

Leaky ReLU is an improved version of ReLU.

Instead of returning **0** for all negative inputs, it allows a small negative output. This helps prevent neurons from becoming permanently inactive (Dying ReLU).

```python
# Create Leaky ReLU activation

leaky_relu = nn.LeakyReLU(negative_slope=0.01)

# Apply activation

y_leaky = leaky_relu(x)

# Plot

plt.figure(figsize=(6,4))
plt.plot(x.numpy(), y_leaky.numpy())
plt.title("Leaky ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

### Observation

- Output Range: **(-∞, ∞)**
- Similar to ReLU
- Allows a small gradient for negative inputs
- Helps solve the Dying ReLU problem

## GELU Activation Function

The Gaussian Error Linear Unit (GELU) is a smooth activation function used in modern Transformer models.

Unlike ReLU, GELU gradually activates neurons instead of making a hard decision.

It is widely used in models such as **BERT**, **GPT**, and **Vision Transformers (ViT)**.

```python
# Create GELU activation

gelu = nn.GELU()

# Apply activation

y_gelu = gelu(x)

# Plot

plt.figure(figsize=(6,4))
plt.plot(x.numpy(), y_gelu.numpy())
plt.title("GELU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

### Observation

- Output Range: **(-∞, ∞)**
- Smooth activation
- Performs better than ReLU in many Transformer architectures
- Commonly used in Large Language Models (LLMs)

```python
# Plot all activation functions together

plt.figure(figsize=(10,6))

plt.plot(x.numpy(), y_sigmoid.numpy(), label="Sigmoid")
plt.plot(x.numpy(), y_tanh.numpy(), label="Tanh")
plt.plot(x.numpy(), y_relu.numpy(), label="ReLU")
plt.plot(x.numpy(), y_leaky.numpy(), label="Leaky ReLU")
plt.plot(x.numpy(), y_gelu.numpy(), label="GELU")

plt.title("Comparison of Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid(True)

plt.show()
```

## Summary

| Activation Function | Output Range | Common Use |
|---------------------|--------------|------------|
| Sigmoid | 0 to 1 | Binary Classification Output Layer |
| Tanh | -1 to 1 | Hidden Layers (Older Networks) |
| ReLU | 0 to ∞ | Hidden Layers (Default Choice) |
| Leaky ReLU | (-∞, ∞) | Avoid Dying ReLU |
| GELU | (-∞, ∞) | Transformers (GPT, BERT, ViT) |
